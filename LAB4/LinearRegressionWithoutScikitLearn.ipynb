```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample dataset
# (You can modify this or load from file if needed)
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])  # Ideally y = 2x
```

```python
# Implement Linear Regression using Gradient Descent

def gradient_descent(X, y, lr=0.01, epochs=1000):
    m = 0  # slope
    b = 0  # intercept
    n = len(X)

    for _ in range(epochs):
        y_pred = m * X + b
        error = y - y_pred

        # Derivatives
        dm = (-2/n) * sum(X * error)
        db = (-2/n) * sum(error)

        # Update parameters
        m -= lr * dm
        b -= lr * db

    return m, b

# Train the model using gradient descent
m_gd, b_gd = gradient_descent(X, y)
print(f"Gradient Descent Results: Slope = {m_gd:.4f}, Intercept = {b_gd:.4f}")
```

```python
# Visualize the results
y_pred_gd = m_gd * X + b_gd
plt.scatter(X, y, label="Data Points")
plt.plot(X, y_pred_gd, color='red', label="Gradient Descent Line")
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
# Least Squares Method (Closed Form)
def least_squares(X, y):
    n = len(X)
    mean_x = np.mean(X)
    mean_y = np.mean(y)

    numerator = sum((X - mean_x) * (y - mean_y))
    denominator = sum((X - mean_x)**2)

    m = numerator / denominator
    b = mean_y - m * mean_x
    return m, b
```

```python
# Train the model using least squares
m_ls, b_ls = least_squares(X, y)
print(f"Least Squares Results: Slope = {m_ls:.4f}, Intercept = {b_ls:.4f}")

# Visualize the results
y_pred_ls = m_ls * X + b_ls
plt.scatter(X, y, label="Data Points")
plt.plot(X, y_pred_ls, color='green', label="Least Squares Line")
plt.title("Linear Regression using Least Squares")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
